```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import r2_score
```

```python
for i in range(1,100):
    X_train,X_test,Y_train,Y_test=train_test_split(principal_x,Y,test_size=0.30,random_state=i)
    lr.fit(X_train,Y_train)
    pred_train=lr.predict(X_train)
    pred_test=lr.predict(X_test)
    
    print(f"At random state {i}, the training accuracy is : {r2_score(Y_train,pred_train)}")
    print(f"At random state {i}, the testing accuracy is : {r2_score(Y_test,pred_test)}")
    print("\n")
```

```python
#At random state 88, the training accuracy is : 0.7737153972651983
#At random state 88, the testing accuracy is : 0.8650195597067971
X_train,X_test,Y_train,Y_test=train_test_split(principal_x,Y,test_size=0.30,random_state=88)
lr.fit(X_train,Y_train)
pred_train=lr.predict(X_train)
pred_test=lr.predict(X_test)
    
print(f"At random state 88, the training accuracy is : {r2_score(Y_train,pred_train)}")
print(f"At random state 88, the testing accuracy is : {r2_score(Y_test,pred_test)}")

```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor

model=[DecisionTreeRegressor(),KNeighborsRegressor(),AdaBoostRegressor(),GradientBoostingRegressor()]
max_r2=0
for rstate in range(40,90):
    X_train,X_test,Y_train,Y_test=train_test_split(principal_x,Y,test_size=0.33,random_state=rstate)
    for i in model:
        i.fit(X_train,Y_train)
        pred_train=i.predict(X_train)
        pred_test=i.predict(X_test)
        r2=r2_score(Y_test,pred_test)
        print("R2 score correspond to Random state", rstate,"is", r2)
        if r2>max_r2:
                    max_r2=r2
                    finalstate=rstate
                    finalmodel=i
print("Max R2score correspond to randomstate", finalstate, "is", max_r2, "and the model is",finalmodel )
        
    
```

```python
#Max R2score correspond to randomstate 84 is 0.8920485507692786 and the model is GradientBoostingRegressor()
X_train,X_test,Y_train,Y_test=train_test_split(principal_x,Y,test_size=0.33,random_state=84)
gdr=GradientBoostingRegressor()
gdr.fit(X_train,Y_train)
pred_train=gdr.predict(X_train)
pred_test=gdr.predict(X_test)
r2=r2_score(Y_test,pred_test)
print("R2 score is", r2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error 
import pandas as pd
MAE= mean_absolute_error(Y_test,pred_test)
print("Mean Absolute Error:",MAE)

MSE=mean_squared_error(Y_test,pred_test)
print("Mean Squared Error:",MSE)

print("Root Mean Squared Error:", np.sqrt(mean_squared_error(Y_test,pred_test)))
```
